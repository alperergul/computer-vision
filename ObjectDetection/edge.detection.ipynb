```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../DATA/sammy_face.jpg')
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(img, threshold1=127, threshold2=127)
```

```python
plt.imshow(edges)
```

```python
med_val = np.median(img)
```

```python
med_val
```

```python
# LOWER THRESHOLD TO EITHER 0 OR 70% OF THE MEDIAN VALUE WHICEVER IS GREATER
lower = int(max(0,0.7*med_val))

# UPPER THRESHOLD TO EITHER 130% OF THE MEDIAN OR THE MAX 255
upper = int(min(255, 1.3*med_val))
```

```python
edeges= cv2.Canny(img, threshold1=lower, threshold2=upper+100)
plt.imshow(edeges)
```

```python
upper
```

```python
blurred_img = cv2.blur(img,ksize=(7,7))
```

```python
edges = cv2.Canny(image=blurred_img, threshold1=lower, threshold2=upper+50)
plt.imshow(edges)
```

```python
plt.imshow(edges, cmap='gray')
```
