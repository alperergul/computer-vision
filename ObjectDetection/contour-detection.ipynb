```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../DATA/internal_external.png")
```

```python
plt.imshow(img, cmap='gray')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
contours, hierarchy = cv2.findContours(img, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

```python
contours
```

```python
external_counters = np.zeros(img.shape)
```

```python
hierarchy
```

```python
for i in range(len(contours)):
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(external_counters, contours, i,255, -1)
```

```python
plt.imshow(external_counters, cmap='gray')
```

```python
internal_counters = np.zeros(img.shape)
for i in range(len(contours)):
    if hierarchy[0][i][3] != -1:
        cv2.drawContours(internal_counters, contours, i,255, -1)
```

```python
plt.imshow(internal_counters ,cmap='gray')
```

```python
smile_counter = np.zeros(img.shape)

for i in range(len(contours)):
    if(hierarchy[0][i][3] == 0):
        cv2.drawContours(smile_counter, contours, i, 255, -1)
```

```python
plt.imshow(smile_counter, cmap='gray')
```
